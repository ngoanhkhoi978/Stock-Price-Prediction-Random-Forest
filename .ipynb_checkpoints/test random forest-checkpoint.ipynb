```python
from datetime import datetime
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import matplotlib.ticker as ticker
from sklearn.metrics import mean_absolute_error, mean_squared_error

from model.RandomForest import RandomForest

TICKER = 'HON'
START_DATE = '1990-01-01'

feature_columns = [
    'Close', 'Returns', 'Volatility', 'Volume_MA7',
    'RSI', 'MACD', 'Oil_Change', 'Stock_Oil_Corr',
    'Vol_x_Oil', 'USD_Change', 'VIX_Change'
]

def directional_accuracy(y_true, y_pred):
    return np.mean(
        np.sign(y_true[1:] - y_true[:-1]) ==
        np.sign(y_pred[1:] - y_pred[:-1])
    )

def evaluate_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    da = directional_accuracy(y_true, y_pred)

    return {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'MAPE': mape,
        'DA':  da
    }


def download_stock_data(ticker='XOM', start_date='2015-01-01', end_date=None):
    if end_date is None:
        end_date = datetime.now().strftime('%Y-%m-%d')

    df_stock = yf. download(ticker, start=start_date, auto_adjust=True)

    macro_tickers = ["CL=F", "DX-Y.NYB", "^VIX", "^TNX"]
    df_macro = yf.download(macro_tickers, start=start_date, auto_adjust=True)

    if df_stock.empty:
        raise ValueError(f"Không thể tải dữ liệu cho mã {ticker}")

    if isinstance(df_stock. columns, pd.MultiIndex):
        df_stock. columns = df_stock.columns.get_level_values(0)

    if isinstance(df_macro.columns, pd. MultiIndex):
        df_macro = df_macro['Close']
    else:
        df_macro = df_macro[['Close']]

    rename_map = {
        'CL=F': 'Oil_Close',
        'DX-Y.NYB': 'USD_Close',
        '^VIX':  'VIX_Close',
        '^TNX':  'TNX_Close'
    }
    df_macro = df_macro.rename(columns=rename_map)

    df_final = df_stock.join(df_macro)
    df_final = df_final. ffill()
    df_final = df_final.reset_index()

    return df_final

def create_features(df):
    data = df.copy()

    data['Returns'] = data['Close'].pct_change()
    data['Volatility'] = data['Returns'].rolling(window=21).std()
    data['Volume_MA7'] = data['Volume'].rolling(window=7).mean()

    # RSI
    delta = data['Close']. diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    data['RSI'] = 100 - (100 / (1 + rs))

    # MACD
    exp1 = data['Close'].ewm(span=12, adjust=False).mean()
    exp2 = data['Close']. ewm(span=26, adjust=False).mean()
    data['MACD'] = exp1 - exp2

    # Macro features
    data['Oil_Change'] = data['Oil_Close'].pct_change()
    data['Stock_Oil_Corr'] = data['Returns'].rolling(window=21).corr(data['Oil_Change'])
    data['Vol_x_Oil'] = data['Volume'] * data['Oil_Change']
    data['USD_Change'] = data['USD_Close'].pct_change()
    data['VIX_Change'] = data['VIX_Close'].pct_change()

    data = data.dropna()
    return data

def prepare_diff_data(df, feature_cols):
    df = df.copy()

    df['Target_Diff'] = df['Close'].shift(-1) - df['Close']
    df = df.dropna()

    X = df[feature_cols].values
    y = df['Target_Diff'].values
    current_prices = df['Close'].values
    dates = df['Date']. values

    split = int(len(X) * 0.85)

    return (X[: split], X[split:],
            y[:split], y[split:],
            current_prices[split:],
            dates[split:])

def visualize_one_step_model(y_true, y_pred, dates, model_name='Model'):
    sns.set(style="whitegrid")

    fig, axes = plt. subplots(2, 1, figsize=(14, 10))

    axes[0].plot(dates, y_true, label='Actual', color='blue', linewidth=1.5)
    axes[0].plot(dates, y_pred, label='Predicted', color='red', linewidth=1.5, alpha=0.7)
    axes[0].set_title(f'{model_name} - Actual vs Predicted Prices')
    axes[0].set_ylabel('Price ($)')
    axes[0].legend(loc='upper left')
    axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    axes[0].xaxis.set_major_locator(ticker.MaxNLocator(nbins=10))

    error = y_true - y_pred
    axes[1].plot(dates, error, label='Error', color='green', linewidth=1)
    axes[1].axhline(y=0, color='black', linestyle='--', linewidth=0.8)
    axes[1].set_title(f'{model_name} - Prediction Error')
    axes[1].set_ylabel('Error ($)')
    axes[1]. set_xlabel('Date')
    axes[1].legend(loc='upper left')
    axes[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    axes[1].xaxis.set_major_locator(ticker.MaxNLocator(nbins=10))

    fig.autofmt_xdate(rotation=45)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    print("Đang tải dữ liệu...")
    df = download_stock_data(TICKER, START_DATE)
    print(f"Đã tải {len(df)} dòng dữ liệu")

    print("Đang tạo features...")
    df_features = create_features(df)
    print(f"Số lượng samples sau khi tạo features: {len(df_features)}")

    print("Đang chuẩn bị dữ liệu...")
    X_train_rf, X_test_rf, y_train_rf_diff, y_test_rf_diff, price_base_test, dates_test = \
        prepare_diff_data(df_features, feature_columns)

    print(f"Train size: {len(X_train_rf)}, Test size: {len(X_test_rf)}")

    print("Đang train Random Forest...")
    rf = RandomForest(
        n_trees=300,
        max_depth=10,
        min_samples_split=2,
        random_state=42
    )
    rf.fit(X_train_rf, y_train_rf_diff)
    print("Đã train xong!")

    print("Đang dự đoán...")
    rf_pred_diff = rf. predict(X_test_rf)

    y_pred_final = price_base_test + rf_pred_diff
    y_true_final = price_base_test + y_test_rf_diff

    metrics_rf = evaluate_metrics(y_true_final, y_pred_final)

    print("\n" + "="*50)
    print("KẾT QUẢ ĐÁNH GIÁ - Random Forest (Diff Strategy)")
    print("="*50)
    for metric, value in metrics_rf.items():
        if metric == 'MAPE':
            print(f"{metric}:  {value:.2f}%")
        elif metric == 'DA':
            print(f"{metric}: {value:.2%}")
        else:
            print(f"{metric}: {value:.4f}")

    visualize_one_step_model(
        y_true=y_true_final,
        y_pred=y_pred_final,
        dates=pd.to_datetime(dates_test),
        model_name='Random Forest (Diff Strategy)'
    )
```
