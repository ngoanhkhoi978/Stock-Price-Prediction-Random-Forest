```python
from datetime import datetime
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ticker = "AAPL"
start_date = "2020-01-01"
end_date = "2025-12-31"
df = yf.download(ticker, start="2020-01-01", end="2025-12-31")
df.columns = df.columns.droplevel(1)
print(df)

df.describe()

plt.figure(figsize=(10,4))
plt.plot(df.index, df['Close'])
plt.title(f'Gi ng ca c phiu {ticker}')
plt.xlabel('Thi gian')
plt.ylabel('Gi')
plt.grid(True)
plt.show()
```

```python
# --- 2. B SUNG D LIU V M (VIX, SP500, USD Index) ---
# Ti d liu v m cng khong thi gian
macro_tickers = ['^VIX', '^GSPC', 'DX-Y.NYB']
print("ang ti d liu v m...")
macro_data = yf.download(macro_tickers, start=start_date, end=end_date)['Close']

# i tn ct cho d hiu
macro_data.columns = ['dxy_close', 'sp500_close', 'vix_close']

# Gp vo DataFrame chnh (dng left join  u tin ngy giao dch ca AAPL)
df = df.join(macro_data, how='left')

# X l d liu thiu (do lch mi gi hoc ngy ngh l khc nhau) bng cch ly gi tr ngy trc 
df[['dxy_close', 'sp500_close', 'vix_close']] = df[['dxy_close', 'sp500_close', 'vix_close']].ffill()

# To c trng bin ng cho v m
df['vix_change'] = df['vix_close'].diff()        # Thay i ca ch s s hi
df['sp500_return'] = df['sp500_close'].pct_change() # Li nhun th trng chung
df['dxy_return'] = df['dxy_close'].pct_change()   # Sc mnh ng USD
```

```python


# --- 1. Tnh ton cc c trng c bn (Code c ca bn) ---
df['log_return'] = np.log(df['Close'] / df['Close'].shift(1))
df['volatility_5'] = df['log_return'].rolling(5).std()
df['ma_5'] = df['Close'].rolling(5).mean()
df['ma_10'] = df['Close'].rolling(10).mean()
df['high_low_range'] = (df['High'] - df['Low']) / df['Close']
df['open_close_change'] = (df['Close'] - df['Open']) / df['Open']

# --- 2. B SUNG: Cc ch bo nng cao (Tm l & Dng tin) ---

# A. RSI (Relative Strength Index) - Ch bo S hi & Tham lam
# Chu k thng dng l 14 ngy. RSI > 70: Qu mua, RSI < 30: Qu bn.
delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).ewm(span=14, adjust=False).mean()
loss = (-delta.where(delta < 0, 0)).ewm(span=14, adjust=False).mean()
rs = gain / loss
df['rsi_14'] = 100 - (100 / (1 + rs))

# B. Stochastic Oscillator (%K) - ng lng gi
# So snh gi ng ca vi bin  gi trong 14 ngy.
low_14 = df['Low'].rolling(window=14).min()
high_14 = df['High'].rolling(window=14).max()
df['stoch_k'] = 100 * ((df['Close'] - low_14) / (high_14 - low_14))

# C. Williams %R - im o chiu
# Tng t Stochastic nhng thang o t -100 n 0. > -20 l qu mua, < -80 l qu bn.
df['williams_r'] = ((high_14 - df['Close']) / (high_14 - low_14)) * -100

# D. MACD (Moving Average Convergence Divergence) - Xu hng
# Chnh lch gia EMA 12 v EMA 26.
ema_12 = df['Close'].ewm(span=12, adjust=False).mean()
ema_26 = df['Close'].ewm(span=26, adjust=False).mean()
df['macd'] = ema_12 - ema_26
df['macd_signal'] = df['macd'].ewm(span=9, adjust=False).mean() # ng tn hiu

# E. On Balance Volume (OBV) - Dng tin thng minh
# Tch ly khi lng da trn vic gi tng hay gim.
df['obv'] = (np.sign(df['Close'].diff()) * df['Volume']).fillna(0).cumsum()

# --- 3. X l d liu v to Target ---

# Loi b cc gi tr NaN sinh ra do rolling window (ln nht l 26 ngy ca MACD)
df = df.dropna()

# To target: Bn ang dng log_return ca ngy hm sau
df['target'] = df['log_return'].shift(-1)

# Loi b hng cui cng (v shift(-1) s to ra NaN  dng cui)
df = df.dropna()

# Hin th kt qu
df.tail()
```
